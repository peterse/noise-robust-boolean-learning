```python
import pandas as pd
file_path = "rnn_hamilton_6_choose_4_nbits16_n2000_bf20_seed1234.txt"
df = pd.read_csv(file_path, sep='\s+', engine='python')
```

```python
df.head()
```

```python
# sort db rows by val_acc
df = df.sort_values(by='val_acc', ascending=False)
df.head(20)
```

```python
# sort db rows by val_acc
df = df.sort_values(by='val_acc', ascending=True)
df.head(20)
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# get average of noiseless_val_acc for all db rows with depth=1 and lr=0.001
def slices(header):
    return [df[(df['depth'] == depth) & (df['lr'] == 0.001)][header].to_numpy() for depth in [1, 2, 3]]

fig, ax = plt.subplots(1, 1, figsize=(5, 5))
for i, header in enumerate(["noiseless_val_acc", "train_acc", "val_acc"]):
    color_i = ['r', 'g', 'b'][i]
    y = slices(header)
    x = [1, 2, 3]
    for subx, suby in zip(x, y):
        ax.scatter(subx*np.ones(len(suby)), suby, c=color_i, alpha=0.2, s=5)
# ax.set_title(f"metric={header}")
ax.legend()
```

```python
def training_acc_and_noiseless_validation_acc(depth):
    return df[(df['lr'] == 0.001) & (df['depth'] == depth)][["train_acc", "noiseless_val_acc"]].to_numpy()
    return df[(df['depth'] == depth) & (df['lr'] == 0.001)][["train_acc", "noiseless_val_acc"]].to_numpy()

def training_acc_and_noisy_validation_acc(depth):
    return df[(df['lr'] == 0.001) & (df['depth'] == depth)][["train_acc", "val_acc"]].to_numpy()
```

```python
# get average of noiseless_val_acc for all db rows with depth=1 and lr=0.001

fig, axes = plt.subplots(1, 2, figsize=(10, 5), sharey=True)
colors = ['r', 'g', 'b']
for i, depth in enumerate([1, 2, 3]):
    axes[0].scatter(*training_acc_and_noiseless_validation_acc(depth).T[::-1], s=5, c=colors[i])
    axes[1].scatter(*training_acc_and_noisy_validation_acc(depth).T[::-1], s=5, c=colors[i])
    
axes[0].set_ylabel("training accuracy")
axes[0].set_xlabel("noiseless validation accuracy")
axes[1].set_ylabel("training accuracy")
axes[1].set_xlabel("noisy validation accuracy")
```
