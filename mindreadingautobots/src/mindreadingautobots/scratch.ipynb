```python
import numpy as np
from itertools import product
```

```python
k = 5
p = 0.1
def maj(b):
    return 1 if np.sum(b) > k/2 else 0

out = 0
for x in product([0,1], repeat=k):
    prob_x = 1 / (2 ** k)
    for e in product([0, 1], repeat=k):
        prob_e = p ** sum(e) * (1-p)**(k - sum(e))
        z = np.array(x) ^ np.array(e)
        if maj(z) != maj(x):
            out += prob_e * prob_x
print(out)
```
